```python
cleaned_df = pd.read_csv("../static/data/cleaned_data.csv")
```

```python
cleaned_df.head()
```

```python
# Convert columns with years to numeric
year_columns = cleaned_df.columns[4:]
cleaned_df[year_columns] = cleaned_df[year_columns].apply(pd.to_numeric, errors='coerce')
# Melt the DataFrame
melted_data = pd.melt(cleaned_df, id_vars=['Country Name', 'Country Code', 'Series Name', 'Series Code'], var_name='Year', value_name='Value')
# Extract only the year from the "Year" column
melted_data['Year'] = melted_data['Year'].str.extract('(\d+)')
# Convert the "Year" column to numeric type
melted_data['Year'] = pd.to_numeric(melted_data['Year'], errors='coerce')
# Replace spaces in column names with underscores
melted_data.columns = melted_data.columns.str.replace(' ', '_')
# Display the melted DataFrame
print(melted_data.head())
```
